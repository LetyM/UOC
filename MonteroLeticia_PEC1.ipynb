## Ejercicio 1

Recientemente hemos contratado los servicios de diferentes plataformas de *streaming*. Hemos creado tres pilas para guardar el contenido que nos gustaría ver, en particular, cada una de las **tres pilas** clasifica documentales, películas y series respectivamente. Sin embargo, algunos de los contenidos de esas pilas ya los hemos visto, con lo que también hemos creado **una cola** con los contenidos ya consumidos. 

Nuestro objetivo es generar **una pila** con la *playlist* de los próximos documentales, películas y series que queremos ver y que anteriormente no hemos visto. Para generar esta pila hay que tener en cuenta que:

1. Queremos ir alternando los contenidos según su categoría en el siguiente orden: documentales, películas y series.
2. Si ya hemos visto un contenido de la categoría que toca, saltamos a la siguiente categoría (v. g. si toca una serie y ya la hemos visto, añadiremos a la pila un documental).
3. No queremos ver contenidos de la misma categoría de forma consecutiva (siempre que podamos evitarlo).
4. Queremos mantener el orden de los contenidos tal como los guardamos en cada una de las pilas.


**Nota:** si tenéis dudas sobre la disposición final de la pila del ejercicio, podéis consultar [este enlace](https://docs.google.com/spreadsheets/d/1dbfVtLYPJcbcUfy9CZo805bpmVpQIlnOKXtbG3ntXVI/edit?usp=sharing) o bien en el PDF `Ejercicio1.pdf` proporcionado junto con la PEC, donde encontraréis una descripción gráfica del estado inicial de las tres pilas y la cola, así como el estado final en el que tiene que quedar la pila. 

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1 punto)**

```python
# The following varibles contain the stacks of documentaties,
# series and movies, as well as the queue with the watched contents

documentaries_stack = ["Social Dilemma", "Wild Wild Country",
                       "Alpha Go", "Chasing Ice", "The Drawn Wall",
                       "Challenger: The Final Flight", "Fyre",
                       "Joan Didion: The Center Will not Hold", "Icaro",
                       "13th"]
series_stack = ["The Wire", "The Sopranos", "Chernobyl", "Big Little lies",
                "True Detective", "Breaking Bad", "Friends", "Black Mirror",
                "Peaky Blinders", "Mad Men"]
movies_stack = ["Dogville", "Persépolis", "Farenheit 451", "Take Shelter",
                "Carol", "Manhattan", "Seven", "The Ring", "Us",
                "Jurassik Park"]
watched_queue = ["Social Dilemma", "Chernobyl", "Take Shelter", "The Sopranos",
                 "Mad Men", "Friends", "Seven", "Fyre"]
```

```python
# Solución

# Creamos un diccionario con el orden que queremos seguir para formar la pila
order_dict = dict()

# Introducimos las listas en el diccionario: documentales, peliculas y series
order_dict["1"] = documentaries_stack
order_dict["2"] = movies_stack
order_dict["3"] = series_stack

# Inicializamos la pila "playlist"
playlist_stack = []

# Mientras haya items en alguna de las pilas, los añadimos a nuestra playlist.

# Para ello, recorremos "order_dict" alternando entre las pilas.
# Si quedan items en la pila que toca y si no hemos visto el que corresponde
# => lo añadimos a la playlist
# En cualquier caso, lo quitamos de su pila.
while max([len(order_dict[k]) for k in order_dict.keys()]) > 0:
    for orderkey in order_dict.keys():
        if len(order_dict[orderkey]) > 0:
            if order_dict[orderkey][-1] not in watched_queue:
                playlist_stack.append(order_dict[orderkey][-1])
            order_dict[orderkey].pop()

# La lista que obtenemos no está en el orden de la pila que queremos,
# la revertimos.
playlist_stack = list(reversed(playlist_stack))


# Finalmente, utilizamos la funcion vista en la unidad 1 para pintar la pila
def print_stack(s, msg_bef=None, msg_after=None):
    if msg_bef:
        print(msg_bef)

    if len(s) == 0:
        max_len = 10
    else:
        max_len = max([len(e) for e in s]) + 4

    print("|" + " "*(max_len+2) + "|")
    for e in s[::-1]:
        print("| " + e + " "*(max_len-len(e)) + " |")
    print("|" + "_"*(max_len+2) + "|")

    print(msg_after+"\n" if msg_after else "\n")


print_stack(playlist_stack, "La pila playlist es:")
```

## Ejercicio 5

Basándonos en las listas de años y meses anteriores:

1. Encontrad cuál es el primer año en el que se lanzó un cohete.
2. Cuántos cohetes se lanzaron desde el primer lanzamiento hasta finales de los años 60s (31/12/1969).
3. ¿Cuál es el año en el que se han lanzado menos cohetes?
4. ¿Cuál es el mes en el que se han lazando más cohetes?

Mostrad el resultado por pantalla debidamente formateado (usando el método `format()`), incluyendo un mensaje explicativo de los valores que mostráis.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1 punto)**

```python
# Solución
from collections import defaultdict
import calendar

# 1. Encontrad cuál es el primer año en el que se lanzó un cohete.
print("El primer año en el que se lanzo un cohete fue: {}"
      .format(min(years_list)))

# 2. Cuántos cohetes se lanzaron desde el primer lanzamiento hasta
# finales de los años 60s

# Nos quedamos con una lista de los años correspondientes y
# calculamos su longitud
years60_list = [year for year in years_list if year <= 1969]
print("El número de cohetes lanzados hasta finales de los 60s es: {}"
      .format(len(years60_list)))

# 3. ¿Cuál es el año en el que se han lanzado menos cohetes?

# Primero creamos un diccionario con una lista de años por año
years_dict = defaultdict(list)
for year in years_list:
    years_dict[year].append(year)

# Ahora creamos un diccionario con el numero de veces que aparece
# el año
len_years_dict = {year: len(year_list) for year, year_list
                  in years_dict.items()}

# Mostramos el año en el que menos lanzamientos ha habido
min_launch_year = min(len_years_dict, key=len_years_dict.get)
print("El año en el que menos cohetes se lanzaron es: {}"
      .format(min_launch_year))

# 4. ¿Cuál es el mes en el que se han lazando más cohetes?

# Procedemos de forma similar para el máximo de los meses

# Primero creamos un diccionario con una lista de meses por mes
months_dict = defaultdict(list)
for month in months_list:
    months_dict[month].append(month)

# Ahora creamos un diccionario con el numero de veces que aparece
# el mes
len_months_dict = {month: len(month_list) for month, month_list
                   in months_dict.items()}

# Mostramos el año en el que menos lanzamientos ha habido
max_launch_month = max(len_months_dict, key=len_months_dict.get)
print("El mes en el que más cohetes se han lanzado es el {} ({})"
      .format(max_launch_month, calendar.month_name[max_launch_month]))
```

## Ejercicio 6

Calculad la diferencia, en días, desde la primera misión espacial de **RVSN USSR** y la primera misión espacial de **US Navy**.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1 punto)**

## Ejercicio 7

Usando **expresiones regulares**:

1. Encontrad todos los cohetes cuya primera palabra del nombre empieza por "S" y acaba por "k"
2. Encontrad todos los cohetes cuyo nombre contiene la palabra "Apollo"

**Nota**: El nom del coet es troba en el camp `detail`.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1 punto)**

```python
# Solución 7.1
import re

# Vemos que la columna "detail" es de tipo Serie
print("La columna 'detail' es de tipo: {}"
      .format(type((space_data['detail']))))

# La pasamos a lista para trabajar más cómodamente
detail = list(space_data['detail'])

# Vamos a buscar los cohetes que empiezan por S y acaban por k
# en el registro del detail.
# Entendemos que para que sea una palabra, debe haber un espacio delante
# y otro detrás.
# También consideramos '-' detrás de la k para coger cohetes numerados
Sk_rockets = list()
for d in range(0, len(detail)):
    if len(re.findall(r" S\w*k[ -]", detail[d])) > 0:
        Sk_rockets.append(detail[d])

print("\nHemos encontrado {} registros en los que hay una palabra "
      "que empienza por S y acaba por k.".format(len(Sk_rockets)))
print("\nLos cuales son: \n{}".format(Sk_rockets))
```

```python
# Solución 7.2

# Procedemos de forma similar para los que contienen "Apollo"

Apollo_rockets = list()
for d in range(0, len(detail)):
    if len(re.findall(r"Apollo", detail[d])) > 0:
        Apollo_rockets.append(detail[d])

print("\nHemos encontrado {} registros que contienen 'Apollo'."
      .format(len(Apollo_rockets)))
print("\nLos cuales son: \n{}".format(Apollo_rockets))
```

Cread un diccionario ordenado (`OrderedDict`) con los siguientes campos:

- company_name
- country (extraed el país del campo `location`)
- date (extraed el valor del campo `datum`)
- status_rocket
- rocket_cost
- status_mission

El diccionario debe de estar indexado por el campo **id** (v. g. 0, 1, 2, ...) y ordenado por el campo **date**. Además solo se deben guardar los elementos que tengan un valor de `rocket_cost` válido. La estructura debe de ser {id0: dict_con_datos_de_id0, id1: dict_con_datos_de_id1, ...}. Por ejemplo, el `dict_con_datos_de_id0` sería: 
                
                (4020,
              {'company_name': 'US Air Force',
               'date': Timestamp('1964-09-01 15:00:00+0000', tz='UTC'),
               'status_rocket': 'StatusRetired',
               'detail': 'Titan IIIA | Transtage 1',
               'rocket_cost': 63.23,
               'country': 'USA',
               'status_mission': 'Failure'}),

**Pista**: podéis usar el método `.to_dict()` de una columna del dataset para convertirla en un diccionario.

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **(1 punto)**

```python
# Solución
from collections import OrderedDict

# Nos quedamos con las columnas con las que vamos a trabajar
ej8_data = space_data[['id', 'company_name', 'datum', 'status_rocket',
                       'detail', 'rocket_cost', 'location', 'status_mission']]

# Formateamos las fechas
ej8_data['datum'] = pd.to_datetime(ej8_data['datum'], utc=True)

# Ordenamos por fecha
ej8_data = ej8_data.sort_values(by='datum')

# Nos quedamos con los registros con un coste de cohete válido
ej8_data = ej8_data[ej8_data['rocket_cost'].notna()]

# Reseteamos los indices para recorrerlos en nuestro diccionario ordenado
ej8_data = ej8_data.reset_index(drop=True)

# Tomamos el pais de la localizacion
country_list = [ej8_data['location'].str.split(", ")[i][-1]
                for i in range(0, len(ej8_data['location']))]
ej8_data['country'] = country_list

# Renombramos la columna "datum" y quitamos "location"
ej8_data = ej8_data.rename(columns={"datum": "date"})
ej8_data = ej8_data.drop(columns="location")

# Ahora vamos a crear el diccionario ordenado
# -Pasamos a diccionario los datos que tenemos
ej8_data_dict = ej8_data.to_dict()

# -Creamos una lista con la cabecera para movernos por ella
head_list = list(ej8_data.head())
head_list.remove('id')

# -Inicializamos el diccionario y lo vamos completando
ej8_dict = OrderedDict()
ej8_dict = {ej8_data_dict['id'][i]:
            {head: ej8_data_dict[head][i] for head in head_list}
            for i in range(0, len(ej8_data))}

# Finalmente, mostramos la primera entrada como ejemplo
print("Mostramos un ejemplo del diccionario ordenado: \n{}"
      .format(list(ej8_dict.items())[0]))
```

La función debe de estar preparada para trabajar con diccionarios muy grandes. Por lo tanto, **se debe evitar guardar los valores individuales mostrados por pantalla en una lista**.

Llamad a la función definida con el diccionario del ejercicio anterior como parámetro.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1 punto)**

```python
# Solución

# Definimos la función
def retiredUSAcost(orddict):
    # Inicializamos las variables que vamos a utilizar para el máximo
    # y para la media (total y conteo)
    max_value = 0
    total_value = 0
    count_value = 0

    # Recorremos el diccionario comprobando las condiciones.
    # Si el registro las cumple, mostramos por pantalla el coste
    # y actualizamos tanto el conteo como la suma de costes
    for ky in orddict.keys():
        if orddict[ky]['country'] == 'USA' and \
           orddict[ky]['status_rocket'] == 'StatusRetired':
            print("El coste de {} fue de {}".
                  format(ky, orddict[ky]['rocket_cost']))
            count_value += 1
            total_value += orddict[ky]['rocket_cost']

            # Si el coste del registro actual es mayor que el max
            # que teníamos guardado, lo actualizamos
            if orddict[ky]['rocket_cost'] > max_value:
                max_value = orddict[ky]['rocket_cost']

    # Calculamos la media
    mean_value = total_value / count_value

    # Devolvemos el máximo y la media
    return max_value, mean_value


# Finalmente, llamamos a la función y mostramos los valores pedidos
print("\nLos resultados son: \nValor maximo: {} \nMedia: {}"
      .format(retiredUSAcost(ej8_dict)[0], retiredUSAcost(ej8_dict)[1]))
```

Recordad que se pueden utilizar librerías científicas para realizar los cálculos de forma sencilla. 

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1 punto)**

```python
# Solución
import pandas as p
import datetime
import statistics

# 1. Coste medio de los lanzamientos llevados a cabo por SpaceX
# desde 01/01/2015 hasta la actualidad y cuyo status de la misión
# ha sido satisfactorio.

ap1 = statistics.mean(
    [ej8_dict[k]['rocket_cost'] for k in ej8_dict.keys()
     if ej8_dict[k]['company_name'] == 'SpaceX'
     and ej8_dict[k]['date'] >= p.to_datetime("2015-01-01", utc=True)
     and ej8_dict[k]['date'] < p.to_datetime(datetime.datetime.now(), utc=True)
     and ej8_dict[k]['status_mission'] == 'Success'])

print("La media pedida en el apartado 1 es: {}".format(ap1))

# 2. Coste medio de los lanzamientos llevados a cabo por SpaceX
# con identificador menor a 523 y cuyo status de la misión
# no ha sido satisfactorio.

ap2 = statistics.mean(
    [ej8_dict[k]['rocket_cost'] for k in ej8_dict.keys()
     if ej8_dict[k]['company_name'] == 'SpaceX'
     and k < 523
     and ej8_dict[k]['status_mission'] == 'Failure'])

print("La media pedida en el apartado 2 es: {}".format(ap2))
```
